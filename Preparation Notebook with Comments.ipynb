### General Prepocessing

#### Prepocessing few continuous variables

During the preprocessing stage, its important to examine the data types of columns, handle missing values, identify and address outliers, and normalize or scale continuous variables as necessary to prepare the data for analysis or modeling. 

For example we expected that "emp_lenght" and "term" columns' data types should be numeric but these are shown as object in the table above.

```python
loan_data['emp_length'].unique()
```

```python
loan_data['emp_length_int'] = loan_data['emp_length'].str.replace('+ years', '', regex=False)
loan_data['emp_length_int'] = loan_data['emp_length_int'].str.replace('< 1 years', str(0), regex=False)
loan_data['emp_length_int'] = loan_data['emp_length_int'].str.replace('n/a', str(0), regex=False)
loan_data['emp_length_int'] = loan_data['emp_length_int'].str.replace(' years', '', regex=False)
loan_data['emp_length_int'] = loan_data['emp_length_int'].str.replace(' year', '', regex=False)
loan_data['emp_length_int'] = loan_data['emp_length_int'].str.replace('< 1', str(0), regex=False)
loan_data['emp_length_int'] = loan_data['emp_length_int'].str.replace('nan', str(0), regex=False)
```

```python
loan_data['emp_length_int'] = pd.to_numeric(loan_data['emp_length_int'], errors='coerce')
```

There are similar issue in "earliest_cr_line" and "loan_issue_date". These are string variables but they are actually supposed to be date variables.

```python
loan_data['earliest_cr_line']
```

```python
# convert to datetime
loan_data['earliest_cr_line_date'] = pd.to_datetime(loan_data['earliest_cr_line'], format ='%b-%y')
```

```python
type(loan_data['earliest_cr_line_date'][0])
```

```python
loan_data['earliest_cr_line_date'].head(10)
```

```python
# check the difference between the earliest credit line and the current date
pd.to_datetime('2017-12-1') - loan_data['earliest_cr_line_date']
```

```python
pd.options.display.float_format = '{:.3f}'.format
# create duration column in months as mths_since_earliest_cr_line
loan_data['mths_since_earliest_cr_line'] = round(pd.to_numeric((pd.to_datetime('2017-12-1') - loan_data['earliest_cr_line_date']).dt.days // 30))
```

```python
# display descriptive statistics about the data frame
loan_data['mths_since_earliest_cr_line'].describe()
```

The results are not fine. In results,  min duration is negative. It is imposible to have time distance between time and future and the pass.

```python
# display negative values in mths_since_earliest_cr_line cause negative minimum value
loan_data.loc[:, ['earliest_cr_line', 'earliest_cr_line_date', 'mths_since_earliest_cr_line']][loan_data['mths_since_earliest_cr_line'] < 0]
```

earliest_cr_line_date are shown in future (2060s). Probably these should be 1960s. This cause the negetive duration between earliest credit line date and today. Let's fix it.

```python
# Adjust dates greater than 2018-01-01 to the 20th century
loan_data.loc[loan_data['earliest_cr_line_date'] > pd.to_datetime('2018-1-1'), 'earliest_cr_line_date'] = \
    loan_data['earliest_cr_line_date'] - pd.DateOffset(years=100)

# Verify the changes
loan_data['earliest_cr_line_date'].sort_values(ascending=True).head()
```

```python
# Calculate the months since the earliest credit line
loan_data['mths_since_earliest_cr_line'] = (
    (pd.to_datetime('2017-12-01') - loan_data['earliest_cr_line_date']).dt.days // 30
)

# Verify the results
loan_data.loc[:, ['earliest_cr_line', 'earliest_cr_line_date', 'mths_since_earliest_cr_line']].head(10)
```

```python
loan_data['mths_since_earliest_cr_line'].describe()
```

The descriptive values are correct. We've imputed the values successfully and our variable is ready for analysis.
